### 1. 背景
为什么python语言适合大数据分析？嗯，我也在思考。目前的理解是，语言机制上提供了几十个双下划线（__）开头和结尾的函数供程序员重载（重新定义和编写内容），使得程序员自定义到数据结构（pandas、numpy等）可以很容易的进行切片、遍历、甚至代数计算。实际上这些类似运算符重载的机制c++也有，但是c++过于复杂且容易出错（对于菜鸟来说吧），所以，吸取了适合大数据科学计算的一些优秀机制后，python开始逆袭了（实际上它的速度还是比较慢的）……

[numpy](https://numpy.org/)的目标就是在底层用c语言的速度实现python的科学计算，但是**它的缺点是多维数组没有名称，只有记住维度的顺序**，这常常给程序员带来困扰，不记得第几维代表的具体含义了。我就是这样的人，特别是不同语言的多维数组常常相反。例如FORTRAN语言中经常用的二维数组是row major还是column major（python也有c还是fortran形式的存放顺序），就是指最左边的维度在内存中是连续还是最右边的维度是连续的。[xarray](http://xarray.pydata.org/en/stable/)通过对维度、坐标与数据进行绑定，带来不一样的世界。

以前用[GrADS](http://cola.gmu.edu/grads/)的时候，一个二进制文件需要一个[ctl](http://cola.gmu.edu/grads/gadoc/descriptorfile.html)文件来描述二进制文件的维数，经向、纬向多少网格，垂向多少网格，时间多少网格等等。NetCDF则把这个信息集成到了二进制文件中，称之为头信息（header）。好处是一个数据一个文件，self-descriptive，并且不同维数的变量也能存一起。从这里已经隐约看到把维度坐标绑定到数据上的好处。

**[xarray](http://xarray.pydata.org/en/stable/)把这一点进行到了极致**。底层是[numpy](https://numpy.org/)，但是维度有名称，比如时间time，这样可以在做平均的时候，API设计成`data_mean = data.mean(dim='time')`，表明沿着时间维度做平均，这样肯定比[numpy](https://numpy.org/)的`data_mean = data.mean(dim=dim_0)`要好得多，因为很快我们就不记得`dim_0`是经度还是时间还是层次。用[xarray](http://xarray.pydata.org/en/stable/)的话两个月后或者两年后你来看代码还是知道自己在做什么（不可否认我经常不知道自己的代码做了啥），实现无需注释的程序代码就是最好的代码。

### 2. 绑定维度坐标带来的好处
大气海洋科学中的数据通常是多维数组，对应于`xarray.DataArray`，或者[numpy](https://numpy.org/)的`ndarray`。我以前用Java的时候设计过这样的类，本质就是一个四维数组，三维空间加一维时间。但是慢慢发觉不够用，某些情况无法涵盖。比如[GrADS](http://cola.gmu.edu/grads/)里面有第五维ensemble，特别是现在集合预报流行了以后，多了这样一维。再比如我现在做的等值线分析，多了一维等值线空间。这些都无法用传统的四维空间描述，因此，[numpy](https://numpy.org/)的`ndarray`和`xarray.DataArray`就具有更广泛的代表性。

[xarray](http://xarray.pydata.org/en/stable/)绑定了数据和维度坐标后，带来了几方面的好处，下面分别简单介绍。

##### 2.1 画图
用过[Matlab](https://www.mathworks.com/)的人都知道画一张温度平面图一般要提供三个参数，经度、纬度、温度，然后用`pcolor(lon, lat, air)`出图。本质上这就是没有做temp上绑定lon、lat坐标的不便之处。绑定后的python代码就是`air.plot()`。下面小试牛刀：

````python
import xarray as xr

# 打开xarray自带的气温数据
dset = xr.tutorial.open_dataset('air_temperature')

air = dset['air']

print(air)

# 画第一个时刻的平面图
air[0].plot()
````

从打印的信息可以清楚看到，`air`是一个`xarray.DataArray`，并且有三维，分别是时间、纬度和经度，具体坐标也都打印了一部分。画图也十分简单，这得益于`DataArray`已经绑定了坐标信息。

##### 2.2 用差分求导数
大气海洋计算经常要用到差分法计算变量的偏导数。比如$\partial u/\partial x$或者$\partial_x u$，表示$u$沿$x$方向的偏导数（上个例子中x就是经度lon）。由于绑定了坐标信息，计算代码就一句话（这里的差分结果单位是K/经度，要转换成K/m需要额外处理）：

````python
# 方法 1, 中心差分法
dTdx1 = air.differentiate('lon')
# 方法 2, 结果少一个元素
dTdx2 = air.diff('lon') / air.lon.diff('lon')

import matplotlib.pyplot as plt

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(11,4))

dTdx1[0].plot(ax=ax[0]) # 画第一个时刻，对应以下左图
dTdx2[0].plot(ax=ax[1]) # 画第一个时刻，对应以下右图
````

两种差分方法结果基本一致，细微的差别来自差分方案。方案1是中心差分，方案2是单向差分（如果认为结果定义在经度lon的半网格上，也可以认为是中央差分，具体参见[交错网格](https://db0nus869y26v.cloudfront.net/en/Arakawa_grids)）。

### 4. 总结
Python是支持特定的运算符重载（[参见双下划线开头的函数](https://rszalski.github.io/magicmethods/#:~:text=%20A%20Guide%20to%20Python%27s%20Magic%20Methods%20,the%20biggest%20advantages%20of%20using%20Python%27s...%20More%20)）。这确实赋予了自定义数据结构在科学计算上拥有简单的语法。比如在Python中，风矢量的大小计算可以写成`speed = sqrt(u**2 + v**2)`, 其中`speed`，`u`，`v`是自定义的多维数组数据结构，这样无需注释一眼就能看明白代码的含义。但在不支持运算符重载的语言中，比如`Java`，只能在多维数组数据结构`Variable`上绑定几个方法：
``` Java
public class Variable{
    // ...
    Variable add(Variable v);
    Variable multiply(Variable v);
}
```
然后同样的计算写为
``` Java
Variable u = ...;
Variable v = ...;
Variable speed = sqrt(u.multiply(u).add(v.multiply(v)))
```
可见Python确实可以用很少的代码写清楚计算逻辑，这也是为什么Python更适合科学计算的一个原因。但是我们也不难看出，对于科学计算，特别是微积分和矢量运算，常规的四则运算符和比较运算符重载已经远远不够。我猜想会不会有更丰富的运算符可以用符号实现而不是函数实现（比如上面例子中的开根运算仍然用函数完成）。当然首先要能用ASCII码定义出统一的微积分符号，而且差分的格式也需要考虑...好吧，想得有点多了。
